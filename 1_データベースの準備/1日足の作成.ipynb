```python
from datetime import datetime
from time import sleep
import requests
import numpy as np
import pandas as pd
```

```python
def t2s(the_time):
    return datetime.fromtimestamp(int(the_time)).strftime('%Y/%m/%d %H:%M')
```

# MySQL 中の1分足を取得

```python
import mysql.connector

cnt = mysql.connector.connect(
    host='localhost',
    port='3306',
    db='bitmex',
    user='bitmex',
    password='bitmex',
    charset='utf8'
)

cur = cnt.cursor()
```

```python
sql = "select t, o, h, l, c, v from xbtusd_1m order by t asc"
cur.execute(sql)
data = cur.fetchall()
```

```python
t2s(data[0][0]), t2s(data[4][0])
```

```python
d = pd.DataFrame(data)
```

```python
d.rename(columns={0:"t", 1:"o", 2:"h", 3:"l", 4:"c", 5:"v"})
```

# １日足の作成

```python
w = 60*24
data2 = []
for i in range(0, int(len(data)/w)*w, w):
    t = data[i][0]
    o = data[i][1]
    h = data[i][2]
    l = data[i][3]
    v = data[i][5]
    for m in range(w):
        h = max(h, data[i+m][2])
        l = min(l, data[i+m][3])
        v += data[i+m][5]
    c = data[i+w-1][4]
    data2.append( (t, o, h, l, c, v) )    
```
